```
#Start with loading all necessary libraries
import numpy as np
#Numpy provides fast mathematical function processing
import pandas as pd
#Pandas is used for easy indexing and data collection tools
from PIL import Image
#Pillow is a very good Python Image library
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
#I am importing necessary packages for Data Visualizations in Python
import matplotlib.pyplot as plt
% matplotlib inline
```

```
#Second, I manually add words from my Resume 
text=("Matt Matt Matt Matt Matt Matt Matt Matt Matt Matt Sirkis Sirkis Sirkis Sirkis Sirkis Sirkis Sirkis Sirks Sirkis Sirkis Text Mining Text Mining Text Mining"  )
```

```
#Third I create the wordcloud object
wordcloud = WordCloud(width=512, height=480, margin=0).generate(text)
```

```
 
#Fourth I insert the code to display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.margins(x=30, y=0)
plt.show()
```
