```
#?Wordcloud


#import world cloud libraries
#STOPWORDS is a function that automatically removes words from your results
#Most commonly imported as a seperate file and joined with the data frame
from wordcloud import WordCloud, STOPWORDS
wordcloud=WordCloud(
                        background_color= "white",
                        stopwords=STOPWORDS,
                        max_words=150,
                        max_font_size=70,
                        random_state=0
                       ).generate(str(Wordclouddata2))

# Display the generated image:
plt.imshow(wordcloud, interpolation= 'bilinear')
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```
